```python
import pyarrow.orc as orc
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import optuna
import joblib

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
from sklearn.compose import ColumnTransformer, TransformedTargetRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score, recall_score, average_precision_score, classification_report
from sklearn.impute import SimpleImputer
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. Define the file path
path = r'C:/BigData/ProjectStaging/dataExport/fraud_mart_final.orc'

# 2. Load the file using PyArrow first (more efficient)
orc_file = orc.ORCFile(path)
table = orc_file.read()

# 3. Convert to Pandas with "self_destruct"
# self_destruct=True: Frees the Arrow memory AS it creates the Pandas DF
# split_blocks=True: Allows data to be scattered in RAM (doesn't need one giant 4GB block)
df = table.to_pandas(split_blocks=True, self_destruct=True)

print(df.head())
print(df.info()) # Check memory usage
```

### TIME EXTRACTION & SORTING

```python
print("Step 1: extracting timestamps and sorting...")

# 1. Extract Unix Timestamp from transaction_id (Format: User-Card-Timestamp)
# taking the last part of the string '61-5-1514771820' -> '1514771820'
df['unix_time'] = df['transaction_id'].str.split('-').str[-1].astype(int)

# 2. Convert to Datetime object
df['datetime'] = pd.to_datetime(df['unix_time'], unit='s')

# 3. Sort by Time (Oldest -> Newest) to respect Time Series nature
df = df.sort_values('datetime').reset_index(drop=True)

df.head(5)
```

### FEATURE ENGINEERING (Time-Aware)

```python
print("Step 2: Engineering features...")

df['amount'] = df['amount'].astype(float)

# 1. Temporal Features
df['minute'] = df['datetime'].dt.minute
df['day_of_week'] = df['datetime'].dt.dayofweek

# 2. Velocity Feature: Transactions in last 24 Hours
# We use a rolling window. This requires setting the timestamp as the index temporarily.
print("   Calculating velocity (transactions last 24h)...")
df = df.set_index('datetime') 
# Count 'transaction_id' (or any column) per user over a 24h window
df['trans_last_24h'] = df.groupby('user_id')['transaction_id'].rolling('24h').count().values
df = df.reset_index() # Reset index back to normal

# 3. Ratio Feature: Amount vs. Median
# We calculate the user's median spending to find anomalies
user_median = df.groupby('user_id')['amount'].transform('median')
df['amount_ratio'] = df['amount'] / user_median

#4. Transaction ammount Difference
df['diff_from_median'] = df['amount'] - df.groupby('user_id')['amount'].transform('median')

#5 mcc Risk 
df['mcc_fraud_rate'] = df.groupby('merchant_category')['label_is_fraud'].transform('mean')

#6. Calculate time difference in seconds between current and previous transaction for each user
df['time_since_last_trans'] = df.groupby('user_id')['unix_time'].diff().fillna(0)

#7. Create a "Rapid Fire" flag (e.g., transactions within 10 seconds)
df['rapid_fire_trans'] = (df['time_since_last_trans'] < 10).astype(int)

# Convert Hour to Circle (Sin/Cos) so that 11pm and 12 am are considered nearby instead of far apart
df['hour_sin'] = np.sin(2 * np.pi * df['hour_of_day'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour_of_day'] / 24)

# Convert Day to Circle so that sunday and monday are considered near to each other instead of far apart
df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)

# Drop the original linear columns to prevent noise
df.drop(['merchant_id', 'card_brand', 'user_id', 'card', 'transaction_id'], axis = 1,errors='ignore', inplace = True)
df.drop(['hour_of_day', 'day_of_week', 'minute'], axis=1, inplace=True)
```

### TIME-BASED SPLIT (Train / Val / Test)

```python
print("Step 3: Splitting data chronologically...")

# Define Split Ratios (70% Train, 15% Val, 15% Test)
n = len(df)
train_end = int(n * 0.70)
val_end = int(n * 0.85)

# Slice the sorted dataframe
train_raw = df.iloc[:train_end]
val_df = df.iloc[train_end:val_end]
test_df = df.iloc[val_end:]

print(f"   Train Range: {train_raw['datetime'].min()} -> {train_raw['datetime'].max()}")
print(f"   Val Range:   {val_df['datetime'].min()} -> {val_df['datetime'].max()}")
print(f"   Test Range:  {test_df['datetime'].min()} -> {test_df['datetime'].max()}")
```

### UNDERSAMPLE & SHUFFLE

```python
print("Step 4: Balancing the Training set...")

# 1. Separate Classes in Training Data
fraud_train = train_raw[train_raw['label_is_fraud'] == 1]
normal_train = train_raw[train_raw['label_is_fraud'] == 0]

# 2. Undersample Normal Class, A 10:1 ratio
ratio = 10 
if len(normal_train) > len(fraud_train) * ratio:
    normal_sample = normal_train.sample(n=len(fraud_train) * ratio, random_state=42)
else:
    normal_sample = normal_train # Keep all if we don't have enough to undersample

# 3. Concatenate and SHUFFLE
train_df = pd.concat([fraud_train, normal_sample], axis=0)
train_df = train_df.sample(frac=1, random_state=42).reset_index(drop=True)

print("\n--- Final Dataset Shapes ---")
print(f"Training Set (Balanced): {train_df.shape} (Fraud Cases: {len(fraud_train)})")
print(f"Validation Set (Raw):    {val_df.shape}")
print(f"Test Set (Raw):          {test_df.shape}")
```

### Finding best XGBoost model for the fraud detection problem, by doing hyper parameter tuning using optuna and bayesian search

```python
# 1. Setup Feature Lists (Same as before)
numeric_features = [
    'amount', 'credit_utilization', 'current_age', 'yearly_income_person', 
    'fico_score', 'trans_last_24h', 'amount_ratio', 'state_mismatch',
    'mcc_fraud_rate', 'diff_from_median', 'time_since_last_trans',
    'hour_sin', 'hour_cos', 'day_sin', 'day_cos', 'rapid_fire_trans'
]
categorical_features = ['use_chip', 'card_type', 'error_code', 'merchant_category']
all_features = numeric_features + categorical_features

# Prepare Data
X_train = train_df[all_features]
y_train = train_df['label_is_fraud']
X_val = val_df[all_features]
y_val = val_df['label_is_fraud']

# 2. Define the Preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline([('imp', SimpleImputer(strategy='median')), ('std', StandardScaler())]), numeric_features),
        ('cat', Pipeline([('imp', SimpleImputer(strategy='constant', fill_value='missing')), 
                          ('enc', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))]), categorical_features)
    ]
)

# 3. Define the Optuna Objective Function
def objective(trial):
    # Suggest hyperparameters
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        # we let Optuna find the best weight
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 1, 50),
        'tree_method': 'hist',
        'random_state': 42,
        'n_jobs': -1
    }

    # Create Pipeline with current parameters
    model_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', xgb.XGBClassifier(**params))
    ])

    # Fit on training data
    model_pipeline.fit(X_train, y_train)

    # Predict on validation data
    preds = model_pipeline.predict_proba(X_val)
    
    # We want to maximize Recall
    score = average_precision_score(y_val, preds[:,1])

    return score

# 4. Run the Study
print("Starting Optuna Optimization for AUPRC...")
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=30, show_progress_bar=True)

# 5. Output Results
print("\nBest AUPRC Score:", study.best_value)
print("Best Hyperparameters:", study.best_params)
```

### Finding best LightGBM model for the fraud detection problem, by doing hyper parameter tuning using optuna and bayesian search

```python
def lgbm_objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 200, 600),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 31, 256),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 1, 15),
        'tree_method': 'hist',
        'verbose' : -1,
        'random_state': 42
    }

    lgbm_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('clf', lgb.LGBMClassifier(**params))])
    lgbm_pipeline.fit(X_train, y_train)
    
    probs = lgbm_pipeline.predict_proba(X_val)[:, 1]
    return average_precision_score(y_val, probs)

study_lgbm = optuna.create_study(direction='maximize')
study_lgbm.optimize(lgbm_objective, n_trials=30, show_progress_bar=True)

print("\nBest AUPRC Score:", study.best_value)
print("Best Hyperparameters:", study.best_params)
```

### Finding best CatBoost model for the fraud detection problem, by doing hyper parameter tuning using optuna and bayesian search

```python
def cat_objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 200, 600),
        'depth': trial.suggest_int('depth', 4, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 1, 15),
        'allow_writing_files': False,
        'verbose': False,
        'random_state': 42,
        # Use GPU if available to speed up 24M rows
        'task_type': 'CPU' 
    }
    
    # 1. Manually transform the data using your existing preprocessor
    # This avoids the Pipeline "is_fitted" check error
    X_train_transformed = preprocessor.fit_transform(X_train)
    X_val_transformed = preprocessor.transform(X_val)
    
    # 2. Initialize and fit the model directly
    clf = CatBoostClassifier(**params)
    clf.fit(X_train_transformed, y_train)
    
    # 3. Get probabilities ([:, 1] for the fraud class)
    probs = clf.predict_proba(X_val_transformed)[:, 1]
    
    return average_precision_score(y_val, probs)

# Run the study as usual
study_cat = optuna.create_study(direction='maximize')
study_cat.optimize(cat_objective, n_trials=30, show_progress_bar=True)

print("\nBest AUPRC Score:", study.best_value)
print("Best Hyperparameters:", study.best_params)
```

### Creating a stacked model using best XGBoost, LGBM and CatBoost to that last ounce of performance

```python
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.utils.validation import check_is_fitted

class CatBoostWrapper(BaseEstimator, ClassifierMixin):
    def __init__(self, iterations=500, depth=6, learning_rate=0.03, scale_pos_weight=1, verbose=False):
        self.iterations = iterations
        self.depth = depth
        self.learning_rate = learning_rate
        self.scale_pos_weight = scale_pos_weight
        self.verbose = verbose
        # This tells sklearn "I am a classifier" without needing __sklearn_tags__
        self._estimator_type = "classifier" 

    def fit(self, X, y):
        self.model_ = CatBoostClassifier(
            iterations=self.iterations,
            depth=self.depth,
            learning_rate=self.learning_rate,
            scale_pos_weight=self.scale_pos_weight,
            verbose=self.verbose,
            allow_writing_files=False
        )
        self.model_.fit(X, y)
        self.classes_ = self.model_.classes_
        return self

    def predict(self, X):
        check_is_fitted(self)
        return self.model_.predict(X)

    def predict_proba(self, X):
        check_is_fitted(self)
        return self.model_.predict_proba(X)
```

```python
# 1. Define your best base models
estimators = [
    ('lgbm', lgb.LGBMClassifier(
        n_estimators=270, max_depth=8, learning_rate=0.11393277130575089, 
        subsample=0.8891521830568911, colsample_bytree=0.8219722945935108, scale_pos_weight=1.5088248590991484
    )),
    ('xgb', xgb.XGBClassifier(
        n_estimators=270, max_depth=8, learning_rate=0.11393277130575089, 
        subsample=0.8891521830568911, colsample_bytree=0.8219722945935108, scale_pos_weight=1.5088248590991484,
        tree_method='hist',verbose=-1
    )),
    ('cat', CatBoostWrapper(
        iterations=270, depth=8, learning_rate=0.11393277130575089, 
         verbose=False, 
        scale_pos_weight=1.5088248590991484 
    ))
]

# 2. Build the Stack
# LogisticRegression is the meta model
final_stack = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    stack_method='predict_proba',
    n_jobs=2
)

# 3. Create the final pipeline
ultimate_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', final_stack)
])

print("Training the Ultimate Ensemble Stack... This might take a while on 24M rows.")
ultimate_pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import precision_recall_curve

X_test = test_df[all_features]
y_test = test_df['label_is_fraud']

# Get probabilities from the ensemble
y_probs = ultimate_pipeline.predict_proba(X_test)[:, 1]

# Get the curve
p, r, thresholds = precision_recall_curve(y_test, y_probs)

# Find threshold for 20% Precision (This is a very strong goal for fraud)
# It means 1 in 5 alerts is a real fraudster.
target_idx = np.argmax(p >= 0.20)
final_thresh = thresholds[target_idx]

print(f"Optimal Threshold: {final_thresh}")
y_final_preds = (y_probs >= final_thresh).astype(int)
print(classification_report(y_test, y_final_preds))
```

```python
# Look for the threshold where Recall is at least 30-40%
# Even if Precision drops to 5-10%
target_recall = 0.35 
idx_balanced = np.where(r >= target_recall)[0][-1]
balanced_thresh = thresholds[idx_balanced]

print(f"Balanced Threshold: {balanced_thresh}")
y_balanced_preds = (y_probs >= balanced_thresh).astype(int)
print(classification_report(y_test, y_balanced_preds))
```

```python
# 1. Get probabilities from your Stacking Ensemble
y_probs = ultimate_pipeline.predict_proba(X_test)[:, 1]

# 2. Define our Risk Tiers
# We define tiers based on probability "certainty"
conditions = [
    (y_probs >= 0.90),
    (y_probs >= 0.50) & (y_probs < 0.90),
    (y_probs >= 0.10) & (y_probs < 0.50),
    (y_probs < 0.10)
]
tiers = ['1. CRITICAL', '2. HIGH', '3. MEDIUM', '4. LOW']

# 3. Create a summary dataframe
results_df = pd.DataFrame({
    'Probability': y_probs,
    'Actual_Fraud': y_test.values,
    'Tier': np.select(conditions, tiers, default = '5. UNKNOWN')
})

# 4. Aggregate the stats per Tier
tier_summary = results_df.groupby('Tier').agg(
    Total_Transactions=('Actual_Fraud', 'count'),
    Actual_Frauds_Found=('Actual_Fraud', 'sum')
).reset_index()

# 5. Calculate key Business Metrics
tier_summary['Fraud_Density (%)'] = (tier_summary['Actual_Frauds_Found'] / tier_summary['Total_Transactions'] * 100).round(2)
tier_summary['%_of_Total_Fraud_Caught'] = (tier_summary['Actual_Frauds_Found'] / y_test.sum() * 100).round(2)

# 6. Add the "Action Plan" (What the bank should actually do)
actions = [
    "Hard Block (Decline)",
    "Friction (Send SMS/2FA)",
    "Passive Monitor (Review Later)",
    "Approve (Trust)"
]
tier_summary['Recommended_Action'] = actions

print("--- STRATEGIC RISK TIERING TABLE ---")
print(tier_summary.to_string(index=False))
```

```python
# 1. Get probabilities
y_probs = ultimate_pipeline.predict_proba(X_test)[:, 1]

# 2. Define Dynamic Percentile Thresholds
# Top 0.1% (Critical), Top 1% (High), Top 5% (Medium)
thresh_critical = np.percentile(y_probs, 99.9)
thresh_high = np.percentile(y_probs, 99)
thresh_medium = np.percentile(y_probs, 95)

print(f"New Thresholds -> Critical: {thresh_critical:.4f}, High: {thresh_high:.4f}, Medium: {thresh_medium:.4f}")

# 3. Apply the new buckets
conditions = [
    (y_probs >= thresh_critical),
    (y_probs >= thresh_high) & (y_probs < thresh_critical),
    (y_probs >= thresh_medium) & (y_probs < thresh_high),
    (y_probs < thresh_medium)
]
tiers = ['1. CRITICAL (Top 0.1%)', '2. HIGH (Top 1%)', '3. MEDIUM (Top 5%)', '4. LOW (Safe)']

results_df = pd.DataFrame({
    'Actual_Fraud': y_test.values,
    'Tier': np.select(conditions, tiers, default='4. LOW (Safe)')
})

# 4. Generate Final Report
tier_summary = results_df.groupby('Tier').agg(
    Total_Trans=('Actual_Fraud', 'count'),
    Frauds_Caught=('Actual_Fraud', 'sum')
).reset_index()

tier_summary['Fraud_Rate (%)'] = (tier_summary['Frauds_Caught'] / tier_summary['Total_Trans'] * 100).round(2)
tier_summary['%_Recall'] = (tier_summary['Frauds_Caught'] / y_test.sum() * 100).round(2)

print("\n--- FINAL CAPSTONE PERFORMANCE ---")
print(tier_summary.to_string(index=False))
```

```python
# Sort by Tier to ensure logical order
tier_summary = tier_summary.sort_values('Tier')

# Calculate Cumulative Recall
tier_summary['Cumulative_Recall (%)'] = tier_summary['%_Recall'].cumsum().round(2)

# Calculate Cumulative Transactions (%)
tier_summary['Cumulative_Workload (%)'] = (tier_summary['Total_Trans'].cumsum() / tier_summary['Total_Trans'].sum() * 100).round(2)

print("\n--- EXECUTIVE SUMMARY TABLE ---")
print(tier_summary[['Tier', 'Fraud_Rate (%)', 'Cumulative_Workload (%)', 'Cumulative_Recall (%)']].to_string(index=False))
```
